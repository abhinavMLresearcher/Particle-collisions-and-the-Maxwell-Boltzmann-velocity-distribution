```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.patches import Circle, Rectangle
from IPython.display import HTML
import random

def collision1D(x0,y0,vx0,vy0,dt):
    vx1=vx0
    vy1=vy0
    x1=vx0*dt+x0
    y1=vy0*dt+y0
    for i in range(len(x0)):
        for j in range(len(x0)):

            if (np.sqrt((x1[i]-x1[j])**2+(y1[i]-y1[j])**2)<=d) and (x1[i]>-1) and (x1[i]<1) and (y1[i]>-1) and (y1[i]<1) and (j>i):
                R=np.array([x1[j]-x1[i],y1[j]-y1[i]])

                n=R/(np.sqrt(sum(R**2)))
                t=np.array([-n[1],n[0]])
                x_u=np.array([1,0])
                y_u=np.array([0,1])

                vi=np.array([vx1[i],vy1[i]])
                vin=np.dot(n,vi)*n
                vit=np.dot(t,vi)*t

                vj=np.array([vx1[j],vy1[j]])
                vjn=np.dot(n,vj)*n
                vjt=np.dot(t,vj)*t
                vin,vjn=vjn,vin


                vx1[i]=vin[0]+vit[0]
                vy1[i]=vin[1]+vit[1]

                vx1[j]=vjn[0]+vjt[0]
                vy1[j]=vjn[1]+vjt[1]

                x1[i]=vx1[i]*dt+x0[i]
                y1[i]=vy1[i]*dt+y0[i]
                x1[j]=vx1[j]*dt+x0[j]
                y1[j]=vy1[j]*dt+y0[j]

            elif (abs(x1[i])>1-r):
                 vx1[i]=-vx1[i]
                 x1[i]=vx1[i]*dt+x0[i]

            elif (abs(y1[i])>1-r):
                 vy1[i]=-vy1[i]
                 y1[i]=vy1[i]*dt+y0[i]
    return x1,y1,vx1,vy1

particles=100
y0=np.array([random.uniform(-1,1) for i in range(particles)])
x0=np.array([random.uniform(-1,1) for i in range(particles)])
vy0=np.ones_like(x0)*0.01#vx0.copy() #np.array([random.uniform(-10, 10) for i in range(particles)])*2
vx0=-np.ones_like(x0)*100*3
dt=0.0005
r=0.05
d=2*r
t_frames=200
x=[x0]
y=[y0]
vx=[vx0]
vy=[vy0]
t=[0]

for i in range(t_frames):
    t.append(t[-1]+dt)
    col=collision1D(x[-1].copy(),y[-1].copy(),vx[-1].copy(),vy[-1].copy(),dt)
    x.append(col[0])
    y.append(col[1])
    vx.append(col[2])
    vy.append(col[3])


import matplotlib.colors as mcolors
import scipy.optimize as opt
from scipy.optimize import curve_fit
colors= mcolors.CSS4_COLORS
list(colors.keys())
fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(12,6))
plt.suptitle("Particle collisions and the Maxwell–Boltzmann velocity distribution")

def update(frame):
    ax1.cla()
    ax2.cla()

    for i in range(particles):
        cer1=Circle((x[frame][i],y[frame][i]),radius=d/2, color=list(colors.keys())[i%148])
        ax1.add_patch(cer1)
    v=np.sqrt(vx[frame]**2+vy[frame]**2)
    #ax2.hist(v, bins=20, range=(0,1200))
    def gaussion(x,sigma,mu):
        return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-mu)**2/(2*sigma**2))
    freq,edge_int=np.histogram(v, bins=10)
    mid_int=np.array([edge_int[i]+(edge_int[i+1]-edge_int[i])/2 for i in range(len(edge_int)-1)])
    freq_N=freq/max(freq)
    mid_int_N=mid_int/max(mid_int)
    popt, pcov = opt.curve_fit(gaussion, mid_int_N,freq_N)
    freq_g_N = gaussion(mid_int_N, popt[0],popt[1])
    freq_g=freq_g_N*max(freq)
    ax2.plot(mid_int,freq_g,color="blue")
    ax2.hist(v,color="red", bins=10, edgecolor="skyblue")
    '''
    cer1=Circle((x[frame][0],y[frame][0]),radius=d/2, color="red")
    cer2=Circle((x[frame][1],y[frame][1]),radius=d/2,color="blue")

    cer3=Circle((x[frame][2],y[frame][2]),radius=d/2,color="yellow")

    cer4=Circle((x[frame][3],y[frame][3]),radius=d/2,color="pink")
    cer5=Circle((x[frame][4],y[frame][4]),radius=d/2,color="black")


    ax1.add_patch(cer1)
    ax1.add_patch(cer2)

    ax1.add_patch(cer3)

    ax1.add_patch(cer4)
    ax1.add_patch(cer5)
    '''
    ax1.set_xlim(-1,1)
    ax1.set_ylim(-1,1)
    ax1.set_xlabel("x", fontsize="19")
    ax1.set_ylabel("y", fontsize="19")
    ax2.set_xlabel("Velcoty (v)", fontsize="19")
    ax2.set_ylabel("The number of Particles", fontsize="19")
    ax2.set_xlim(0,800)
    ax2.set_ylim(0,30)
    return
ani = FuncAnimation(
    fig,
    update,
    frames=t_frames,
    interval=100,
    blit=False
)
HTML(ani.to_jshtml())
```
